Ниже мы создали фрейм, в котором у нас содержится информация о урожае, который мы высадили , в данных есть название культуры и результат посева.  

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, color="orchid")
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="red")
plt.title("Failure Rates")
plt.ylabel("Number Of Dead Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db")
plt.bar(plants, germinated, width=0.4, color="#2ecc71", align="edge")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated, label="Failed Seedlings")
plt.bar(plants, germinated, color="#2ecc71", label="Germinated Seedlings")
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title("Greenhouse Seedlings")
```

Вот тут, например, мы можем увидеть, сколько саженцов у нас взошло для каждой из имеющихся у нас культур. Например, самая высокая урожайность у нас произошла для репы, ее мы получим больше всего в этом в году.

```python
plt.figure(figsize=(6,6))
plt.barh(plants, germinated, color="#3498db")
```

```python
titanic = pd.read_csv("data/titanic.csv")
titan = sns.load_dataset("titanic")
```

```python
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")
```

```python
plt.hist(titanic["age"])
plt.show()
```

Мы можем определить, что самым частым возрастом на Титанике можно счиатать от 20 до 25 лет. Это очень молодые люди.

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
second_class = titanic[titanic["pclass"] == 2]["age"]
third_class = titanic[titanic["pclass"] == 3]["age"]
```

```python
plt.hist(first_class, label="1st Class", alpha=0.5, color="#ffad01",bins=20)
plt.hist(second_class, label="2nd Class", alpha=0.5, color="#3498db",bins=20)
plt.hist(third_class, label="3rd Class",alpha=0.5, color="#e74c3c",bins=20)
plt.legend()
```

Теперь мы можем также увидеть, что чаще всего люди такого возраста встречались в 3 классе. А люди от 40 до 60 лет в 1 классе.

Ниже мы создали фрейм из реально произошедших снежных штормов,у нас была информация за последние 24 часа и за следующие 24 часа.

```python
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure(figsize=(8,6), dpi=200)
plt.title("CO Show Storm Report")
plt.bar(resorts, snow_last_24, color="#74b9ff", label="Last 24hrs")
plt.bar(resorts, snow_next_24, bottom=snow_last_24, color="#7ed6df", label="Next 24hrs")
plt.legend()
plt.ylabel("Inches")
```

Мы видим, что ситуация на "Волчьем ручье" ухудшается, и в ближайшие 24 часа они повысятся до 30.

На данном графике мы также отобразили информацию о возрастных категориях, но только отделили ее, чтобы увидеть более точное  распределение по возрастам.

```python
sales = pd.read_csv('data/kc_house_data.csv', parse_dates=['date'])
```

```python
mas = ['Mon', 'Tue','Wed','Thu', 'Fri', 'Sat', 'Sun']
o_sales = sales['date'].dt.month.value_counts().sort_index()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(mas,sales['date'].dt.dayofweek.value_counts())
plt.xticks(fontsize=13)
plt.title("Sales By Week Day")

plt.subplot(1,2,2)
plt.plot(o_sales.index,o_sales)
plt.title("Sales By Month")
plt.xticks(fontsize=13)
plt.xlabel('Month')
plt.ylabel("Home Sales")
plt.show()
```

Мы построили график, отображающий продажу домов по месяцам и по дням недели , мы видим, что квартира чаще продаются весной-летом, в понедельник-вторник.

```python
labels=titan['who'].unique().tolist()
titan[['pclass','who']].value_counts()
labels_to_class = titan[['pclass','who']].value_counts().index
labels_to_class = ['3rd M', '1st M', '3rd W','2nd M','1st W', '2nd W','3rd C', '2nd C', '1st C']
```

```python
plt.figure(figsize=(8,8), facecolor="#E4717A")

colors = ['#8B8378', '#53868B','#2E8B57']
colors_pclass = ['#EEDFCC', '#FFEFDB', '#7AC5CD','#FAEBD7',
                    '#8EE5EE','#98F5FF', 
                  '#50c878','#43CD80','#4EEE94']



bigger = plt.pie(titan['who'].value_counts(), 
                 labels=labels, 
                 colors=colors,
                 startangle=90)
smaller = plt.pie(titan[['pclass','who']].value_counts(),
                  labels=labels_to_class,radius=0.7,
                  startangle=10,
                  labeldistance=1,
                  colors=colors_pclass,
                  autopct='%.1f%%')



plt.title("Distribution by class", size=20)
plt.tight_layout()
plt.show()
```

На данном графике мы видим, что всех категорий людей у нас одинаково больше всего в 3 классе.И также мы видим , что количество людей уменьшается от класса к классу для каждой категории.
