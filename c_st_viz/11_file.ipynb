```python
shots = lagida.groupby("Team").agg(
    total=("Shots", sum),
    on_target=("Shots on target", sum)
)
shots
```

```python
shots["accuracy"] = shots["on_target"] / shots["total"]
```

```python
fig, ax = plt.subplots(2,1, sharex=True)
ax_1 = shots['accuracy'].nlargest(5).sort_values(ascending=False).plot( 
    kind="barh", 
    ax=ax[0],
     color="green",
    title="Most Accurate Teams")
shots['accuracy'].nsmallest(5).sort_values(ascending=False).plot(
    kind="barh", 
    ax=ax[1],
    color="red",
    title="Least Accurate Teams")
plt.xlabel("On Target Percentage")
```
